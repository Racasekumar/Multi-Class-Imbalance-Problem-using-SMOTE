```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=500, n_features=2, n_informative=2,
                          n_redundant=0, n_repeated=0, n_classes=3,
                          n_clusters_per_class=1,weights=[0.05,0.1,0.85],
                          class_sep=0.8, random_state=0)
```

```python
X[:,0]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(8,4))
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y, ax=ax[0])
idx, c = np.unique(y,return_counts=True)
sns.barplot(x=idx, y=c, ax=ax[1])
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
X_resampled, y_resampled = SMOTE(random_state=42).fit_resample(X,y)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,6))
sns.scatterplot(x=X_resampled[:,0], y=X_resampled[:,1], hue=y_resampled, ax=ax[0])
idx, c = np.unique(y_resampled, return_counts=True)
sns.barplot(x=idx, y=c, ax=ax[1])
plt.title("oversampled using SMOTE")
plt.show()
```

```python
print("original sample:", np.unique(y, return_counts=True))
print("after sampling using SMOTE:", np.unique(y_resampled, return_counts=True))
```
